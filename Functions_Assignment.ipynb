Q6 -  Explain the concept of generators in Python and how they are defined?

Ans - Generators in Python are a special type of iterator that allow you to iterate over data lazily, meaning they produce items one at a time and only when required. This makes them more memory-efficient compared to other iterables, like lists, because they don't store all the values in memory at once.

Key Differences Between a Generator and a Normal Function:

yield vs return: A normal function uses return to return a value, but yield is used in a generator function to return a value and pause execution, saving the function’s state for later use.

Memory Efficiency: With normal functions, all results are returned at once and stored in memory (like with a list). Generators produce values one at a time, so they use less memory.

Q7 - What are the advantages of using generators over regular functions?

Ans - Generators provide several distinct advantages over regular functions, especially when it comes to performance, memory efficiency, and handling large datasets. Here are the key benefits:

Memory efficiency (one value at a time, no need to store entire dataset).

Lazy evaluation (compute values only when needed).

Faster processing for large or infinite datasets (no pre-computation).

Infinite sequences without running out of memory.

Cleaner and more concise code compared to using a full iterator class or list.
Automatic state management between iterations.

Q9 -  Explain the purpose and usage of the `map()` function in Python?

Ans - The map() function in Python is used to apply a given function to all items in an iterable (like a list, tuple, etc.) and returns an iterator that yields the results.

Returns an iterator: The result of map() is not a list but an iterator, so you need to convert it to a list (or another data structure) if you want to access the results in memory.

Lazy Evaluation: Since map() returns an iterator, it generates the results lazily, meaning it computes the results on demand rather than all at once.

Multiple Iterables: You can pass more than one iterable to map(), and the function will apply to items in a parallel fashion across all iterables. The function must accept as many arguments as there are iterables.

```python
#Ans
def even_num(numbers):
    return sum(num for num in numbers if num % 2 == 0)

numbers = (1, 2, 3, 4, 5, 6, 7, 8)
result = even_num(numbers)
print(result)
```

Q4 -  Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
#Ans
is_prime = lambda n: n > 1 and all(n % i != 0 for i in range(2, int(n**0.5) + 1))

for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is prime.")
```

Q5 - Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms?

```python
#Ans
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        fib = self.a
        self.a, self.b = self.b, (lambda x, y: x + y)(self.a, self.b)
        self.count += 1
        return fib

# Example usage
fib = FibonacciIterator(8)
for num in fib:
    print(num)
```

Q6 -  Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
#ans
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in powers_of_two(7):
    print(power)
```

Q7 - Implement a generator function that reads a file line by line and yields each line as a string.

```python
#Ans
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return
for line in read_file_line_by_line("your_file.txt"):
    print(line)
```

Q9 - Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
#Ans
celsius_temps = [0, 25, 30, 15, -5, 40]

fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))

print(fahrenheit_temps)
```

Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string?

```python
#Ans
def remove_vowels(input_string):
    vowels = "aeiouAEIOU"
    result = filter(lambda char: char not in vowels, input_string)

    return ''.join(result)

input_string = "Hello, World!"
output_string = remove_vowels(input_string)
print("Original String:", input_string)
print("String without vowels:", output_string)
```
